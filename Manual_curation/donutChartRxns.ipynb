```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
met_pathway=["Fatty acid and lipids biosynthesis", "Cofactor, Carrier and vitamin biosynthesis", "Amino Acid Biosynthesis", "Carbohydrate Biosynthesis", "Nucleoside and nucleotide biosynthesis", "Inorganic Nutrient Metabolism",
             "Secondary Metabolite Biosynthesis", "Amino Acid Degradation",
              "Cellular Respiration", "Photosynthesis", "Glyolysis", "Other"]

number_rxns=[96,95,152,61,66,25,52,60,87,28,58,190]
```

```python
sns.set(font_scale = 0.8)
plt.figure(figsize=(8,8))
# sort the data by the number of reactions in descending order
data = sorted(zip(number_rxns, met_pathway), reverse=True)

# unpack the sorted data into two separate lists
number_rxns, met_pathway = zip(*data)

plt.pie(
    x=number_rxns, labels=met_pathway,
    colors=sns.color_palette('pastel'),
    startangle=90,
    # show percentages
    autopct='%1.2f%%',
    # move the percentage inside the arcs
    pctdistance=0.80,
    # add space between the arcs
    explode=[0.05]*len(met_pathway)
)
hole = plt.Circle((0, 0), 0.65, facecolor='white')
plt.gcf().gca().add_artist(hole)
plt.savefig("UnblockedModelPathways.jpg", dpi=300, bbox_inches='tight')
plt.show()

%cd C:\Users\duart\OneDrive\Documentos\UNIANDES\PHASEOLUS LUNATUS\RESULTS\FINALMODEL_EXPLORE
```
