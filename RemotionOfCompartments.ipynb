```python
import cobra
from cobra.io import load_model
from pathlib import Path
from cobra.io import load_json_model, save_json_model, load_matlab_model, save_matlab_model, read_sbml_model, write_sbml_model
import logging
from cobra import Model, Reaction, Metabolite
```

```python
%cd C:\Users\duart\OneDrive\Documentos\UNIANDES\PHASEOLUS LUNATUS\MODELS\OTHER_SPECIES\Phaseolus vulgaris
modelPvul = read_sbml_model(str("Pvul_CobraTB.xml"))
```

```python
for i in modelPvul.metabolites:
    if i.compartment == 'CCO__45__IN':
        modelPvul.metabolites.get_by_id(i.id).compartment='e'

```

```python
rmvMets=[]
rmvRxns=[]
for j in modelPvul.metabolites:
    if j.compartment == 'e' or j.compartment=='CCO__45__OUT' or j.compartment=='CCO__45__PM__45__BAC__45__NEG' or j.compartment =='CCO__45__MIDDLE':
        rmvMets.append(j)
        for k in j.reactions:
            rmvRxns.append(k)
```

```python
len(rmvMets)
```

```python
len(rmvRxns)
```

```python
modelPvul.remove_metabolites(rmvMets)
modelPvul.remove_reactions(rmvRxns)
```
