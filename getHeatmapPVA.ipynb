```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%cd C:\Users\duart\OneDrive\Documentos\UNIANDES\PHASEOLUS LUNATUS\GENOMIC\Daniela

# Read the input table into a pandas dataframe
df = pd.read_csv('columnsPVAmatrix5.0.csv',sep=";",index_col=0)


# Create the heatmap
fig, ax = plt.subplots(figsize=(12, 18))
cmap = sns.color_palette(['#96D2ED','#0F5257'])


sns.set(font_scale=0.9)
sns.heatmap(df, cmap=cmap, xticklabels=True, cbar_kws={'label': 'Values'}, ax=ax)
plt.savefig('heatmap2.0.jpg', dpi=300, bbox_inches='tight')

```
