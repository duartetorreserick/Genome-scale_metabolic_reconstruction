```python
%cd C:\Users\duart\OneDrive\Documentos\UNIANDES\PHASEOLUS LUNATUS\GENOMIC\Daniela\Pvulgaris
TEvulgaris=pd.read_csv('DBFinal_pvulgarisV1_EDTA_RM_INPACTOR.csv',sep='\t')
```

```python
df2 = pd.read_csv('coordenadasGenes.txt', sep='\t')
```

```python
df2
```

```python

for i in range(len(df2['Start'])):
    # Definir los límites del rango
    chr=df2.loc[i,'Chr']
    end1 = df2.loc[i,'Start']
    start1 = end1-10000
    start2 = df2.loc[i,'End']
    end2 = start2+10000
    
    # # Seleccionar las filas de la primera tabla que caen dentro del rango
    mask1 = (TEvulgaris['Chr']==chr) & (TEvulgaris['Start'] >= start1) & (TEvulgaris['Start'] <= end1)
    mask2 = (TEvulgaris['Chr']==chr) & (TEvulgaris['End'] >= start2) & (TEvulgaris['End'] <= end2)

    res1=TEvulgaris[mask1]
    res2 = TEvulgaris[mask2].rename(columns={"Chr": "Chr_2", "Start": "Start_2", "End": "End_2", "Size": "Size_2", "Taxonomy": "Taxonomy_2"})
    res1=res1.reset_index()
    res2=res2.reset_index()
    result=pd.concat([res1,res2],axis=1)
    print(result)
    with open ('TEvulgaris_10kb.txt', 'a') as f:
         result.to_csv(f, sep='\t', index=False)
        
```

```python
for i in range(len(df2['End'])):
    # Definir los límites del rango
    chr=df2.loc[i,'Chr']
    start = df2.loc[i,'End']
    end = start+10000
    print(chr,start,end)
    
    
    # # Seleccionar las filas de la primera tabla que caen dentro del rango
    mask = (TEvulgaris['Chr']==chr) & (TEvulgaris['End'] >= start) & (TEvulgaris['End'] <= end)
    result = TEvulgaris[mask]
    
    # # Imprimir el resultado
    with open ('TEvulgaris_10kbmore.txt', 'a') as f:
        f.write("DATAFRAME\n")
        result.to_csv(f, sep='\t', index=False)
        
```
